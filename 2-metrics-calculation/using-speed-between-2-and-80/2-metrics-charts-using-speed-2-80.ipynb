# Vehicle Degree

```python
# Vehicle Degree
import matplotlib.pyplot as plt
import pandas as pd

days_to_analyze = [1,5,4,17,12,20]

for day in days_to_analyze:

    df_degree_per_bus_graph = pd.read_parquet(f"../metrics-calculation/using-speed-2-80/connectivity-metrics/connections-degree-per-minute/MO_1510{1}/")\
        .sort_values(by=["time"])

    ax = df_degree_per_bus_graph.plot(x ="time",y = "avg_degree",title=f"Average node degree - {day}/10/2015",figsize=(15,8))
    # x axis label
    ax.set_xlabel("hour",size=20)
    # y axis label
    ax.set_ylabel("Average node degree",size=20)
    
    # title fontsize
    ax.title.set_size(20)
    plt.grid(True)
    # size of the number in both axis
    ax.tick_params(axis='both', which='major', labelsize=15)
    
    ax.get_figure().savefig(f'../charts-results/using-speed-2-80/degree-vertices-per-minute/day-{day}-degree-vehicles.png',bbox_inches='tight')
```

```python
# ECDF connectivy degree
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt


days_to_analyze = [1,5,4,17,12,20]

for day in days_to_analyze:

    df = pd.read_parquet(f"../metrics-calculation/using-speed-2-80/connectivity-metrics/connections-per-vehicle-per-graph/MO_1510{day}/")


    x= np.sort(df["number_connections"])
    y= np.arange(1,len(x)+1) / len(x)
    plt.plot(x,y,marker=".",linestyle="none")
    ax = plt.gca()
    plt.margins(0.02)

    ax.set_xlabel("number of connections",size=15)
    # y axis label
    ax.set_ylabel("ECDF",size=15)

    #title fontsize
    ax.title.set_size(20)
    plt.grid(True)
    # size of the number in both axis
    ax.tick_params(axis='both', which='major', labelsize=15)
    plt.savefig(f"../charts-results/using-speed-2-80/connections-degree-ecdf-day-{day}.png",dpi=300)
    plt.close()
```

```python
# Histogram repeated connections

import matplotlib.pyplot as plt
import pandas as pd

days_to_analyze = [1,5,4,17,12,20]

day = 20

df = pd.read_parquet(f"../metrics-calculation/using-speed-2-80/connectivity-metrics/repeated-connection-per-day/MO_1510{day}/")
ax = df["count_per_day"].plot.hist(logy=True,figsize=(10,10),bins=50)
plt.grid(True)
ax.set_xlabel("repeated connections",size=20)
ax.set_ylabel("frequência",size=20)
plt.title(f"Histogram repeated connections {day}/10/2015",size=20)
plt.savefig(f'../charts-results/using-speed-2-80/historgrama-total-{day}.png',bbox_inches='tight')
plt.close()
```

```python
# ECDF Repeated Connections

import numpy as np

days_to_analyze = [1,5,4,17,12,20]

for day in days_to_analyze:
    
    df = pd.read_parquet(f"../metrics-calculation/using-speed-2-80/connectivity-metrics/repeated-connection-per-day/MO_1510{day}/")

    x = np.sort(df["count_per_day"])
    
    y = np.arange(1,len(x)+1) / len(x)
    
    plt.title(f"ECDF Repeated Connection - {day}/10/2015",size=15)
    
    _ = plt.plot(x,y,marker=".",linestyle="none")
    
    plt.margins(0.02)
    plt.savefig(f'../charts-results/using-speed-2-80/ecdf-repeated-connections-{day}.png',bbox_inches='tight')
    plt.close()
```

# Speed Charts

```python
# average speed per day

import pandas as pd

speed_per_day = pd.read_csv("../metrics-calculation/using-speed-2-80/speed-calculation/statistics/1-speed_per_vehicle_stats.csv")

day_column = [day for day in range(1,32)]
speed_per_day["day"] = day_column

ax = speed_per_day.plot.bar(x ="day",y = "avg_speed",title=f"Average speed per day",figsize=(10,8))
# x axis label
ax.set_xlabel("day",size=15)
# y axis label
ax.set_ylabel("Avg Speed (km/h)",size=15)
# y range
ax.set_ylim(0,21)
# size of the number in both axis
ax.tick_params(axis='both', which='major', labelsize=15)
# for p in ax.patches:
#     ax.annotate(str(round(p.get_height(),2)), (p.get_x() * 1.00, p.get_height() * 1.006),size=13)
# title fontsize
ax.title.set_size(20)

# saving chart as png
ax.get_figure().savefig('../charts-results/using-speed-2-80/avg-speed-per-day.png')
```

```python
# average speed per hour per day
import pandas as pd
import matplotlib.pyplot as plt

for day in range(1,32):

    speed_per_hour = pd.read_parquet(f"../metrics-calculation/using-speed-2-80/speed-calculation/speed-per-hour-per-day/MO_1510{day}/").sort_values(by=["hour_avl"])

    ax = speed_per_hour.plot(x="hour_avl",y ="avg_speed",title=f"Average speed per hour - {day}/10/2015",figsize=(10,8))
    # x axis label
    ax.set_xlabel("hour",size=20)
    # y axis label
    ax.set_ylabel("Avg speed (km/h)",size=20)
    # Setting x axis ticks
#     plt.xticks([hour for hour in range(6,23)])
    # Title size
    ax.title.set_size(20)

    # size of the number in both axis
    ax.tick_params(axis='both', which='major', labelsize=20)
    plt.grid(True)

    # saving chart as png
    ax.get_figure().savefig(f'../charts-results/using-speed-2-80/speed-per-hour-charts/avg-speed-per-hour-day-{day}-10-2015.png',bbox_inches='tight')
    plt.close()
```
