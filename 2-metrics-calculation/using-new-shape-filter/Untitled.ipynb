```python
# Vehicle Degree
import matplotlib.pyplot as plt
import pandas as pd

days_to_analyze = [1,5,4,17,12,20]

for day in days_to_analyze:

    df_degree_per_bus_graph = pd.read_parquet(f"../metrics-calculation/using-new-map-matching-filter/connectivity-metrics/connections-degree-per-minute-one-hour/MO_1510{1}/")\
        .sort_values(by=["time"])

    ax = df_degree_per_bus_graph.plot(x ="time",y = "avg_degree",title=f"Average node degree - {day}/10/2015",figsize=(15,8))
    # x axis label
    ax.set_xlabel("hour",size=20)
    # y axis label
    ax.set_ylabel("Average node degree",size=20)
    
    # title fontsize
    ax.title.set_size(20)
    plt.grid(True)
    # size of the number in both axis
    ax.tick_params(axis='both', which='major', labelsize=15)
    
    ax.get_figure().savefig(f'./teste-day-{day}-degree-vehicles.png',bbox_inches='tight')
```
