```python
sparkSession = pyspark.sql.SparkSession(sc)
```

```python
hadoop_conf = sparkSession._jsc.hadoopConfiguration()
hadoop_conf.set("fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
# hadoop_conf.set("fs.s3a.access.key", AWS_ACCESS_KEY_ID)
# hadoop_conf.set("fs.s3a.secret.key", AWS_SECRET_ACCESS_KEY)
```

```python
# reading a working day file  day 1/10/2015
traces = sparkSession.read.option("header", "false").csv('s3a://mobility-traces-bus-sp/MO_15101.csv')
column_names = "dt_server,dt_avl,id_line,longitude,latitude,id_avl,event,id_point".split(",")
for i in range(len(column_names)):
    traces = traces.withColumnRenamed("_c%s" % i, column_names[i])
```

```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
import pyspark.sql.functions as f
```

```python
traces.registerTempTable("traces")
```

```python
# traces.select("id_avl").distinct
avl_ids = sqlContext.sql("SELECT distinct id_avl FROM traces")
for avl in avl_ids.collect():
    print(avl[0])
```

```python
# Saving data from 2 random buses in 1/10/2015 (working day)
traces_11332 = traces.filter(f.col("id_avl") == "11332")
traces_11332.count()
```

```python
traces_51256 = traces.filter(f.col("id_avl") == "51256")
traces_51256.count()
```

```python
traces_11332.write.csv('s3://mobility-traces-bus-sp/partioned-day-bus/working-day/11332')
traces_51256.write.csv('s3://mobility-traces-bus-sp/partioned-day-bus/working-day/51256')
```

```python
# reading a weekend day 4/10/2015
traces_weekend = sparkSession.read.option("header", "false").csv('s3a://mobility-traces-bus-sp/MO_15104.csv')
column_names = "dt_server,dt_avl,id_line,longitude,latitude,id_avl,event,id_point".split(",")
for i in range(len(column_names)):
    traces_weekend = traces_weekend.withColumnRenamed("_c%s" % i, column_names[i])
```

```python
traces_weekend.registerTempTable("traces_weekend")
```

```python
# traces.select("id_avl").distinct
avl_ids_weekend = sqlContext.sql("SELECT distinct id_avl FROM traces_weekend")
for avl in avl_ids_weekend.collect():
    print(avl[0])
```
