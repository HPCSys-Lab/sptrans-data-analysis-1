```python
# Testing plotting shape matching with traces
import pandas as pd
shapes =  pd.read_csv("./aux-files/shapes.csv")
shapes.head(10)
```

```python
import folium
import pandas as pd

bus = 57116

map_folium = folium.Map([-23.477476,-46.610409], zoom_start=12)

tile = folium.TileLayer('OpenStreetMap').add_to(map_folium)

df = pd.read_parquet(f"./plot-data/bus-{bus}-plotting-shape-matching/")

df_shape_61116 = df.loc[df["shape_id"] == 61116]
shapes_df = shapes.loc[shapes["shape_id"] == 61116]

coords = df_shape_61116[["trace_y","trace_x"]].values.tolist()
coords_shapes = shapes_df[["shape_pt_lat","shape_pt_lon"]].values.tolist()

for coord in coords:
    icon = folium.features.CustomIcon('red-point.png', icon_size=(10,10))
    folium.Marker(coord,icon=icon).add_to(map_folium)

for coord in coords_shapes:
    icon = folium.features.CustomIcon('green-point.png', icon_size=(10,10))
    folium.Marker(coord,icon=icon).add_to(map_folium)
    

map_folium.save(f"./plot-data/shape-matching-with-traces-{bus}-day-1-shape-61116.html")
```
