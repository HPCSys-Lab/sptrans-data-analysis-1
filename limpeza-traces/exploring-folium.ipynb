```python
#https://deparkes.co.uk/2016/06/03/plot-lines-in-folium/
import pandas as pd
import geopandas
import folium

df = pd.read_csv('../MO_1611.txt')
df.columns = ['server','register','line','long','lat','bus','event','nao_sei']
```

```python
busx = df.loc[(df['bus'] == 51661)]
ordered_bus = busx.sort_values(['register'])
bus = ordered_bus[['register','lat','long','event']]
locations = bus[['long','lat']]
locationlist = locations.values.tolist()
map = folium.Map(location=[-23.533773, -46.625290], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point]).add_to(map)

folium.PolyLine(locationlist, color="red", weight=2.5, opacity=1).add_to(map)
```

```python
oneday = bus[bus['register'].str.contains('2016-01-02') & bus['event'] == 0]
locations = oneday[['long','lat']]
locationlist = locations.values.tolist()
map2 = folium.Map(location=[-23.533773, -46.625290], zoom_start=12)
for point in range(0, len(locationlist)):
    # little dots to represent the trajectory instead of the default marker
    # just change the radius
    folium.CircleMarker(locationlist[point],radius=3,color='#3186cc',fill=True,fill_color='#3186cc',fill_opacity=1.0).add_to(map2)

# folium.PolyLine(locationlist, color="red", weight=2.5, opacity=1).add_to(map2)
```
